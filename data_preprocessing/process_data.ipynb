```python
import pandas as pd
import numpy as np
import json
```

```python
descriptions = pd.read_csv('gene_descriptions.csv', index_col=0)
descriptions = descriptions.fillna('')
descriptions = descriptions.set_index(['Gene.stable.ID'])
descriptions = descriptions.drop(columns=['Gene.name','Transcript.stable.ID'])
descriptions
```

```python
go = pd.read_csv('go_terms.csv', index_col=0).fillna('')
go.head()
```

```python
go_terms = go.groupby(by=['Gene.name','Gene.stable.ID'])['GO.term.name'].apply(set).apply(list).reset_index()
```

```python
go_terms['GO.term.name'] = go_terms['GO.term.name'].map(lambda x: list(filter(None, x)))
go_terms = go_terms.set_index(['Gene.stable.ID'])
go_terms
```

```python
counts = pd.read_csv('GSE131230_counts_official.csv', index_col=0)
counts.index = counts.index.set_names(['Gene.stable.ID'])
counts_dict = counts.to_dict('records')
```

```python
counts_dict
```

```python
final = go_terms.join(descriptions, how='outer')
final = final.reset_index()
final
```

```python
counts_dict
```

```python
final['cell_values'] = final['Gene.stable.ID'].map(counts_dict)
final
```

```python
final.to_json('final_data.json',orient='index')
```
